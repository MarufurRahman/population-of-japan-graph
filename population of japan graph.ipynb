```python
# Import panda and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
japan = data[data.country == 'Japan']
```

```python
japan
```

```python
plt.plot(japan.year, japan.population / 10**6)
plt.legend(['Japan'])
plt.title('Population of Japan graph')
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
